```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
import seaborn as sns
import sqlite3
import datetime
```

```python
df2 = pd.read_csv('reddit_body_final.csv')
```

```python
df2['TIMESTAMP'] = pd.to_datetime(df2['TIMESTAMP'])
```

```python
df2['date'] = df2['TIMESTAMP'].dt.date

df2['year'] = df2['TIMESTAMP'].dt.year
```

```python
#modelled data
fig, ax = plt.subplots(figsize=(15,7))
df2['year'] = df2['year'].astype(str)
df2.groupby(['year']).count()[['POST_ID']].plot(ax=ax)
df2.groupby('year').count()[['POST_ID']]
plt.ylabel('number_of_posts')
plt.title('Number of posts per year')
df2.groupby('year').count()[['POST_ID']]
```
