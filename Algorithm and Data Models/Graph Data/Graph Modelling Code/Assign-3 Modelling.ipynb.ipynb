```python
import pandas as p
```

```python
import numpy as np
```

```python
%%time
a1 = p.read_csv('soc-redditHyperlinks-body.tsv', delimiter='\t')
```

```python
%%time
s1 = a1['SOURCE_SUBREDDIT'].unique().tolist()
t1 = a1['TARGET_SUBREDDIT'].unique().tolist()
s1.extend(t1)
vertices = list(dict.fromkeys(s1))
vertices.sort()
len(vertices)
```

```python
#Edge list
```

```python
%%time
edges = list(zip(a1['SOURCE_SUBREDDIT'],a1['TARGET_SUBREDDIT']))
edges.sort()
e1 = list(dict.fromkeys(edges))
len(e1)
e1
```

```python
#Adjacency Matrix n List
```

```python
%%time
v1 = len(vertices)
a2 = np.zeros((v1,v1), dtype=int)
a3 = []
for l in range(v1):
    a3.append([])
#else: 
#     return "Error"
for m in e1:
    row_num = vertices.index(m[0])
    col_num = vertices.index(m[1])
    a2[row_num][col_num]=1
    a3[row_num].append(m[1])
```
